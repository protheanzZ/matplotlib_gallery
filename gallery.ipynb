# 通过style.available 调整图片样式

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
th = np.linspace(0, 2*np.pi, 128)


def demo(sty):
    mpl.style.use(sty)
    fig, ax = plt.subplots(figsize=(3, 3))

    ax.set_title('style: {!r}'.format(sty), color='C0')

    ax.plot(th, np.cos(th), 'C1', label='C1')
    ax.plot(th, np.sin(th), 'C2', label='C2')
    ax.legend()
    

demo('default')
demo('seaborn')
```

```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib import style

styles = style.available
x = np.linspace(-5, 5, 100000)
def styles_demo():
    fig = plt.figure(figsize=(1920/80, 1080/80))
    for i in range(5):
        for j in range(6):
            loc = i*6+j+1
            try:
                style.use(styles[loc-1])
                ax = fig.add_subplot(5, 6, loc)
                ax.plot(x, x, label='liner')
                ax.plot(x, x**2, label='square')
                ax.plot(x, x**3, label='cube')
                ax.plot(x, np.cos(x), label='cos')
                ax.plot(x, x**(1/x), label='$x**(1/x)$')
                ax.legend(loc='best')
                ax.set_ylim(-5, 5)
                ax.set_title('style:%s, num:%s' % (styles[loc-1], loc))
            except IndexError:
                pass
    # 减少输出图形边距
    plt.savefig('26 styles.png', bbox_inches='tight')
```

```python
styles_demo()
```
